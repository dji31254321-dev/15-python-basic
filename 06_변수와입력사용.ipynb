```python
# 변수 여러개를 한번에 만들기
x, y, z = 10, 20, 30
print(x)
print(y)
print(z)
```

```python
x = y = z = 10
print(x)
print(y)
print(z)
```

```python
x, y = 10, 20
x, y = y, x
print(x)
print(y)
```
