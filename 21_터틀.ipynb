```python
# 다각형에 색칠하기
import turtle as t

n = 6
t.shape("turtle")
t.color("red")
t.begin_fill()
for i in range(n):
    t.forward(100)
    t.right(360/5)
t.end_fill()
```

```python
# 원을 반복해서 그리기
import turtle as t
n = 60
t.shape("turtle")
t.speed("fastest")
for i in range(n):
    t.circle(120)
    t.right(360/n)
```

```python
# 선으로 복잡한 무늬 그리기
import turtle as t

t.shape("turtle")
t.speed("fastest")
for i in range(360):
    t.forward(i)
    t.right(91)
```
