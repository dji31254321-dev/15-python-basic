```python
# 반복문으로 딕셔너리의 키-값 쌍을 모두 출력하기
x = {
    "a": 10,
    "b": 20,
    "c": 30,
    "d": 40
}
for i in x:
    print(i, end=" ")
```

```python
for key, item in x.items():
    print(key, item)
```

```python
# 키만 출력하기
for key in x.keys():
    print(key)
```

```python
# 값만 출력하기
for value in x.values():
    print(value)
```

```python
# 딕셔너리 표현식 사용하기(딕셔너리 컴프리헨션)
## 1. {키 : 값 for 키, 값 in 딕셔너리}
## 2. dict({키 : 값 for 키, 값 in 딕셔너리})
keys = ["a", "b", "c", "d"]
x = {key: value for key, value in dict.fromkeys(keys).items()}
print(x)
```

```python
{key: 0 for key in dict.fromkeys(keys).keys()}
```

```python
x = {
    "a": 10,
    "b": 20,
    "c": 30,
    "d": 40
}
{value: 0 for value in x.values()}
```

```python
# 딕셔너리 표현식에서 if 조건문 사용하기
x = {
    "a": 10,
    "b": 20,
    "c": 30,
    "d": 40
}
for key, value in list(x.items()):
    if value == 20:
        del x[key]
print(x)
print(list(x.items()))
```

```python
x = {
    "a": 10,
    "b": 20,
    "c": 30,
    "d": 40
}
{key: value for key, value in x.items() if value != 20}
```

```python
# 딕셔너리의 할당과 복사
x = {"a":0, "b":0, "c":0, "d":0}
y = x
```

```python
x is y
print(id(x))
print(id(y))
```

```python
y["a"] = 99
print(x)
print(y)
```

```python
x = {"a":0, "b":0, "c":0, "d":0}
y = x.copy()
x is y
y['a'] = 99
print(x)
print(y)
```

```python
x = {"a": {"python": "2.7"}, "b": {"python": "3.6"}}
y = x.copy()
y["a"]["python"] = "2.7.15"
print(x)
print(y)
```

```python
import copy
x = {"a": {"python": "2.7"}, "b": {"python": "3.6"}}
y = copy.deepcopy(x)

y["a"]["python"] = "2.7.15"
print(x)
print(y)
```
