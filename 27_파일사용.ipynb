```python
# 파일에 문자열 쓰기
## 파일객체 = open(파일이름, 파일모드)
## 파일객체.write(문자열)
## 파일객체.close()

file = open("./files/hello.txt", "w")
file.write("Hello, World!")
file.close()
```

```python
# 파일에서 문자열 읽기
## 변수 = 파일객체.read()
file = open("files/hello.txt", "r")
s = file.read()
print(s)
file.close()
```

```python
# 자동으로 파일 객체 닫기
## with open(파일이름, 파일모드) as 파일객체:
#       코드      
with open("files/hello.txt","r")as file:
    s = file.read()
    print(s)
```

```python
# 반복문으로 문자열 여러 줄을 파일에 쓰기

with open("files/hello2.txt", "w") as file:
    for i in range(3):
        file.write(f"Hello world {i}\n")
```

```python
# 리스트에 들어있는 문자열을 파일에 쓰기
lines = ["안녕하세요\n", "파이썬\n", "코딩도장입니다.\n"]
with open("files/hello3.txt","w", encoding="utf-8")as file:
    file.writelines(lines)
```

```python
# 파일의 내용을 한 줄씩 리스트로 가져오기
## 변수 = 파일객체.readlines()

with open("files/hello3.txt", "r", encoding = "utf-8") as file:
    lines = file.readlines()
    print(lines)
```

```python
# 파일의 내용을 한 줄씩 읽기
with open("files/hello3.txt", "r", encoding = "utf-8") as file:
    line = None
    while line != "":
        line = file.readline()
        print(line.strip("\n"))
```

```python
# for 반복문을호 파일의 내용을 줄 단위로 읽기
with open("files/hello3.txt", "r", encoding = "utf-8") as file:
    for line in file:
        print(line.strip("\n"))
```

```python
# 파이썬 객체를 파일에 저장하기, 가져오기
import pickle
name = "james"
age = 17
address = "서울시 서초구 반포동"
scores = {
    "korean": 90,
    "english": 95,
    "mathematics": 85,
    "science": 82
}

with open("files/james.p","wb") as file:
    pickle.dump(name, file)
    pickle.dump(age, file)
    pickle.dump(address, file)
    pickle.dump(scores, file)
```

```python
import pickle
with open("files/james.p", "rb") as file:
    name = pickle.load(file)
    age = pickle.load(file)
    address = pickle.load(file)
    scores = pickle.load(file)

    print(name)
    print(age)
    print(address)
    print(scores)
```
