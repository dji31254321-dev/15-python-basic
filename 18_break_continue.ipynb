```python
# break로 반복문 끝내기
i = 0
while True:
    print(i)
    i += 1
    if i == 100:
        break
```

```python
# for에서 break로 반복문 끝내기
for i in range(10000):
    print(i)
    if i == 100:
        break
    
```

```python
# continue로 코드 실행 건너뛰기

for i in range(100):
    if i % 2 == 0:
        continue
    print(i)
```

```python
# while 반복문에서 continue로 코드 실행 건너뛰기
i = 0
while i < 100:
    i += 1
    if i % 2 == 0:
        continue
    print(i)
```

```python
# 입력한 횟수대로 반복하기
count = int(input("숫자"))
i = 0
while True:
    print(i)
    i += 1
    if i == count:
        break
```

```python
# 입력한 숫자까지 홀수 출력하기
count = int(input("숫자"))
for i in range(count+1):
    if i % 2 == 0:
        continue
    print(i)

```

```python
start, stop = map(int, input().split())
i = start
while True:
    if i % 10 == 3:
        i+=1
        continue
    if i > stop:
        break
    print(i, end = ' ')
    i += 1
```
