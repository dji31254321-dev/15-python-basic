```python
# elif 사용하기
x = 20
if x == 10:
    print("10입니다.")
elif x == 20:
    print("20입니다.")
```

```python
#음료수 자판기
button = int(input())
if button == 1:
    print("콜라")
elif button == 2:
    print("사이다")
elif button == 3:
    print("환타")
else:
    print("제공하지 않는 메뉴")
```

```python
x = int(input())
if 11<=x<=20:
    print("11~20")
elif 21<=x<=30:
    print("21~30")
else:
    print("아무것도 해당하지 않음")
```
