```python
for i in range(5):
    for j in range(5):
        print("j: ", j, sep = '', end= ' ')
    print("i: ", i, '\\n', sep='')
```

```python
# 사각형으로 별 출력하기
for i in range(5):
    for j in range(5):
        print("*", end="")
    print()
```

```python
# 사각형 모양 바꾸기
for i in range(3):
    for j in range(7):
        print("*", end="")
    print()
```

```python
# 계단식으로 별 출력하기
for i in range(5):
    for j in range(5):
        if j < i:
            print("*",end="")
    print()
```

```python
# 대각선으로 별 출력하기
for i in range(5):
    for j in range(5):
        if j == i:
            print("*", end="")
    print()
```

```python
for i in range(5):
    for j in range(5):
        if j == i:
            print("*", end="")
        else:
            print(" ", end="")
    print()
```

```python
for i in range(5):
    for j in range(5):
        if i<j:
            print("*",end="")
        else:
            print(" ", end="")
    print()
```
