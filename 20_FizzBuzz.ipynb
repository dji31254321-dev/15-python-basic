```python
# 1부터 100까지 숫ㅈ 출력하기
for i in range(100):
    print(i)
```

```python
# 3의 배수 Fizz, 5의 배수는 Buzz
for i in range(100):
    if i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)
```

```python
# 3과 5의 공배수 처리하기
for i in range(100):
    if i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz")
    elif i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)
```

```python
# 논리연산자를 사용하지 않고 3과 5의 공배수 처리하기
for i in range(100):
    if i % 15 == 0:
        print("FizzBuzz")
    elif i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)
```

```python
# 코드 단축하기
for i in range(101):
    print("Fizz" * (i%3==0) + "Buzz" * (i%5==0) or i)
```
