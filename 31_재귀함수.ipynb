```python
def hello(count):
    if count == 0:
        return
    print("hello world", count)
    count -= 1
    hello(count)
```

```python
hello(5)
```

```python
# 재귀호출로 팩토리얼 구하기
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n-1)
```

```python
factorial(5)
```

```python
def is_palindrome(word):
    if len(word) < 2:
        return True
    if word[0] != word[-1]:
        return False
    return is_palindrome(word[1:-1])
print(is_palindrome('hello'))
print(is_palindrome('level'))
```

```python
def fib(n):
    if n == 0:
        return 0
    if n ==1:
        return 1
    return fib(n-1) + fib(n-2)
```

```python
fib(20)
```
