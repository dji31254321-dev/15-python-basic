```python
x = 10
def foo():
    x = 20
    print(x)

foo()
print(x)
```

```python
# global 전역변수
x = 10
print(id(x))
def foo():
    global x
    print(id(x))
    x = 20
    print(id(x))
    print(x)

foo()
print(x)
```

```python
# 함수 안에서 함수 만들기
# def 함수이름():
#     코드
#     def 함수이름():
#         코드
def print_hello():
    hello = "Hello, world"
    def print_message():
        print(hello)
    print_message()
print_hello()
```

```python
# 지역 변수 변경하기
def A():
    x = 10
    def B():
        x = 20
    B()
    print(x)
A()
```

```python
# nonlocal 지역변수
def A():
    x = 10
    def B():
        nonlocal x
        x = 20
    
    B()
    print(x)
A()
```

```python
# nonlocal이 변수를 찾는 순서
def A():
    x = 10
    y = 100
    
    def B():
        x = 20
        def C():
            nonlocal x
            nonlocal y

            x = x + 30
            y = y + 300
            print(x)
            print(y)
        C()
    B()
A()
```

```python
# global로 전역변수 사용하기
x = 1
def A():
    x = 10
    def B():
        x = 20
        def C():
            global x
            x = x + 30
            print('1', x)
        C()
        print('2',x)
    B()
    print('3',x)
A()
```

```python
# 클로저 사용하기
def calc():
    a = 3
    b = 5
    def mul_add(x):
        return a * x + b
    return mul_add
```

```python
c = calc()
```

```python
c(1), c(2), c(3)
```

```python
list(map(c,[1,2,3,4]))
```

```python
# lambda로 클로저 만들기
def calc():
    a = 3
    b = 5
    return lambda x: a * x + b
c = calc()
print(c(1), c(2), c(3))
```

```python
# 클로저의 지역변수 변경하기
def calc():
    a = 3
    b = 5
    total = 0
    def mul_add(x):
        nonlocal total
        total = total + a * x + b
        print(total)
    return mul_add
```

```python
c = calc()
print(c(1))
print(c(2))
print(c(3))
```
