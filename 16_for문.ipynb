```python
# for와 range 사용하기
# for 변수 in range(횟수):
#     반복할 코드
```

```python
for i in range(100):
    print("hellow world")
```

```python
# 반복문에 변수의 변화 알아보기
for i in range(100):
    print("hello world", i)
```

```python
for i in range(5, 12):
    print("hello world",i)
```

```python
for imsi in range(0,10,2):
    print("hello world", imsi)
```

```python
for a in range(10,0):
    print("hello world", a)
```

```python
list(range(10,0))
```

```python
for a in range(10,0,-1):
    print("hello world", a)
```

```python
print(list(range(10)))
print(list(reversed(range(10))))
```

```python
for c in reversed(range(10)):
    print("hello world", c)
```

```python
# 시퀀스 객체로 반복하기
a = [10, 20, 30, 40, 50]
for imsi in a:
    print(imsi)
```

```python
fruits = ("apple", "orange", "grape")
for fruit in fruits:
    print(fruit)
```

```python
for letter in "Python":
    print(letter)
```

```python
x = [49, -17, 25, 102, 8, 62, 21]
for i in x:
    print(i*10, end=' ')
```

```python
dan = int(input())
for i in range(1, 10):
    print(f"{dan} * {i} = {dan*i}")
```
