```python
# 문자열 판단하기
# re.match("패턴", 문자열)
import re
re.match("Hello", "Hello, world!")
re.match("Python", "Hello, world!")
```

```python
# 문자열이 맨 앞에 오는지 맨 뒤에 오는지 판단하기
# re.search("패턴", "문자열")

print(re.search("^Hello", "Hello, world"))
print(re.search("world$", "Hello, world"))
```

```python
# 범위 판단하기
# [0-9]*
# [0-9]+
re.match("[0-9]*", "1234")
re.match("[0-9]", "1234")
re.match("[0-9]+", "1234")
re.match("[0-9]+", "abcd")
```

```python
print(re.match("a*b", "b"))
print(re.match("a+b", "b"))
print(re.match("a*b", "aab"))
print(re.match("a+b", "aab"))
```

```python
# 문자? -> 문자가 1개 있거나 없을때
# [0-9]? -> 0~9까지의 숫자가 1개 있거나 없을때
# . -> 아무문자 한개
print(re.match("abc?d", "abd"))
print(re.match("abc?d", "abcd"))
print(re.match("ab[0-9]?c", "ab3c"))
print(re.match("ab[0-9]?c", "abc"))
print(re.match("ab.d", "abxd"))
```

```python
# 문자 개수 판단하기
## 문자{개수}
## (문자열){개수}

print(re.match("h{3}", "hhhello"))
print(re.match("h{3}", "hhello"))
print(re.match("(hello){3}", "hellohellohello"))
print(re.match("(hello){3}", "hellohello"))
```

```python
## [0-9]{개수}
print(re.match("[0-9]{3}-[0-9]{4}-[0-9]{4}", "010-2753-3528"))
print(re.match("[0-9]{3}-[0-9]{4}-[0-9]{4}", "010-2753-352"))
```

```python
## (문자){시작개수, 끝개수}
## (문자열){시작개수, 끝개수}
## [0-9]{시작개수, 끝개수}
print(re.match("[0-9]{2,3}-[0-9]{3,4}-[0-9]{4}", "02-100-1000"))
```

```python
## a-z
## A-Z
print(re.match("[a-zA-z0-9]+" , "Hello1234"))
print(re.match("[a-zA-z0-9]" , "Hello1234"))
print(re.match("[A-Z0-9]" , "hello1234"))
```

```python
# 가-힣
print(re.match("[가-힣]+", "홍길동"))
print(re.match("[가-힣]+", "abc홍길동"))
```

```python
# 특정 문자 범위에 포함되지 않는지 판단하기
# [^범위]*
# [^범위]+
print(re.match("[^A-Z]+", "Hello"))
print(re.match("[^A-Z]+", "hello"))
```

```python
## ^[범위]*
## ^[범위]+
print(re.search("^[A-Z]+", "Hello"))
```

```python
## [범위]*$
## [범위]+$
print(re.search("[0-9]+$","hello1234"))
```

```python
# 특수문단 판단하기
## \특수문자

print(re.search("\\*+", "1**2"))
print(re.match("[$()a-zA-Z0-9]+", "$(document)"))
```

```python
# \d -> [0-9]와 같음, 모든 숫자
## \D -> [^0-9]와 같음, 숫자를 제외한 모든 문자
## \w -> [a-zA-Z0-9_] 영문 대소문자, 숫자, 밑줄문자
## \W -> [^a-zA-Z0-9_] 영문 대소문자, 숫자, 밑줄문자를 제외한 모든 문자

print(re.match("\\d+", "1234"))
print(re.match("\\D+", "Hello"))
print(re.match("\\w+", "Hello1234"))
print(re.match("\\W+", "(:)"))
```

```python
# # 공백처리하기
## \s -> \t \n \r \f \v 공백
## \S -> 공백제외
print(re.match("[0-9a-zA-Z ]+", "hello 1234"))
print(re.match("[0-9a-zA-Z\\s]+", "hello 1234"))
```

```python
# 그룹 사용하기
## (정규표현식)(정규표현식)
## 매치객체.group(그룹숫자)
m = re.match("([0-9]+) ([0-9]+)", "10 295")
print(m.group(1))
print(m.group(2))
print(m.group())
print(m.group(0))
print(m.groups())
```

```python
## (?P<이름>정규표현식)
## 매치객체.group(그룹이름)

m = re.match("(?P<func>[a-zA-Z_][a-zA-Z0-9_]+)\\((?P<arg>\\w+)\\)", "print(1234)")
print(m.group("func"))
print(m.group("arg"))
```

```python
## re.findall("패턴", "문자열")
re.findall("[0-9]+","1 2 Fizz 4 Buzz Fizz 7 8")
```

```python
# 문자열 바꾸기
## re.sub("패턴", "바꿀문자열", "문자열", "바꿀횟수")
re.sub("apple|orange", "fruit", "apple box orange tree")
```

```python
re.sub("[0-9]+", "n", "1 2 Fizz 4 Buzz Fizz 7 8")
```

```python
## 교체함수(매치객체)
## re.sub("패턴", 교체함수, "문자열")

def multiple10(m):
    n = int(m.group())
    return str(n*10)

re.sub("[0-9]+", multiple10, "1 2 Fizz 4 Buzz Fizz 7 8")
```

```python
# 찾은 문자열을 결과에 다시 사용하기
# \\숫자

re.sub("([a-z]+) ([0-9]+)", "\\2 \\1 \\2 \\1", "hello 1234")
```
