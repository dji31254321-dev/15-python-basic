```python
# 덧셈 함수 만들기
## def 함수이름(매개변수1, 매개변수2):
##     코드
def add(a, b):
    print(a + b)
```

```python
add(10, 20)
```

```python
# 함수의 결과를 반환하기
def add(a, b):
    return a + b
x = add(10, 20)
print(x)
```

```python
# 함수에서 값을 여러개 반환하기
def add_sub(a, b):
    return a + b, a - b
```

```python
a, b = add_sub(10, 20)
print(a)
print(b)
```

```python
# 함수의 호출 과정 보기
def mul(a, b):
    c = a * b
    return c

def add(a, b):
    c = a + b
    print(c)
    d = mul(a, b)
    print(d)

x = 10
y = 20

add(x, y)
```

```python
x = 10
y = 3
def get_quotient_remainder(a, b):
    return a//b, a%b

quotient, remainder = get_quotient_remainder(x, y)
print("몫: {0}, 나머지: {1}".format(quotient, remainder))
```

```python
x, y = map(int,input().split())
def calc(a, b):
    return a+b, a-b, a*b, a/b
a, s, m, d = calc(x, y)
print("덧셈: {0}, 뺄셈: {1}, 곱셈: {2}, 나눗셈: {3}".format(a, s, m, d))
```
