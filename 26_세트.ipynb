```python
# 세트 만들기
## 세트 = {값1, 값2, 값3}
## 1. 순서가 정해져 있지 않다.
## 2. 중복 될 수 없다.


fruits = {
    "strawberry", "grape", "orange", "pineapple", "cherry", "orange"
}
print(fruits)
```

```python
# 집합연산 사용하기
## 세트1 | 세트2
## set.union(세트1, 세트2)
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}
# 합집합
print(a | b)
print(a.union(b))
print(set.union(a, b))
```

```python
# 교집합 
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

print(a & b)
print(set.intersection(a,b))
print(a.intersection(b))
```

```python
# 차집합
## 세트1 - 세트2
## set.difference(세트1, 세트2)
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}
print(a - b)
print(set.difference(a, b))
print(a.difference(b))

print("="*10)
print(b-a)
print(set.difference(b,a))
print(b.difference(a))

```

```python
# 대칭차집합
## 세트1 ^ 세트2
## set.symmetric_differemce(세트1, 세트2)
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}
print(a ^ b)
print(set.symmetric_difference(a, b))
print(a.symmetric_difference(b))
```

```python
# 집합 연산 후 할당 연산자 사용하기
## 세트1 |= 세트2
## 세트1.update(세트2)
a = {1, 2, 3, 4}
a |= {5}

print(a)

a.update({6})
print(a)
```

```python
## 세트1 &= 세트2
## 세트1.interection_update(세트2)
a = {1, 2, 3, 4}
a &= {0, 1, 2, 3, 4}
print(a)

a = {1, 2, 3, 4}
a.intersection_update({0, 1, 2, 3, 4})
print(a)
```

```python
## 세트1 ^= 세트2
## 세트1.symmetric_difference_update(세트2)
a = {1, 2, 3, 4}
a ^= {3, 4, 5, 6}
print(a)

a = {1, 2, 3, 4}
a.symmetric_difference_update({3, 4, 5, 6})
print(a)
```

```python
a = {1, 2, 3, 4}
a.intersection({3, 4})
a.intersection_update({3, 4})
print(a)
```

```python
# 부분집합과 상위집합 확인하기
## 현재세트 <= 다른세트
## 현재세트.issubset(다른세트)
a = {1, 2, 3, 4}
print(a <= {1, 2, 3, 4})
print(a.issubset({1, 2, 3, 4, 5}))
```

```python
## 현재세트 < 다른세트
a = {1, 2, 3, 4}
a <= {1, 2, 3, 4, 5}
```

```python
## 현재세트 >= 다른세트
## 현재세트.issuperset(다른세트)
a = {1, 2, 3, 4}
print(a >= {1, 2, 3, 4})
print(a.issuperset({1, 2, 3, 4}))
```

```python
## 현재세트 > 다른세트
a = {1, 2, 3, 4}
a > {1, 2, 3}
```

```python
# 세트가 같은지 다른지 확인하기
a = {1, 2, 3, 4}
print(a == {1, 2, 3, 4})
print(a == {4, 2, 1, 3})
```

```python
# 세트가 겹치지 않는지 확인하기
## 현재세트.isdisjoin(다른세트)
## 겹치는 요소가 있으면 False, 없으면 True
a = {1, 2, 3, 4}
print(a.isdisjoint({5, 6, 7, 8}))
print(a.isdisjoint({3, 4, 5, 6}))
```

```python
# 세트 표현식 사용하기
a = {i for i in "apple"}
print(a)
```

```python
# 세트 표현식에 if 조건문 사용
## 1. {식 for 변수 in 세트 if 조건식}
## 2. set(식 for 변수 in 세트 if 조건식)
a = {i for i in "pineapple" if i not in "apl"}
print(a)
```
